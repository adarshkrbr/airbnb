```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
airbnb_df = pd.read_excel(r'C:\Users\admin\Desktop\Maths refresher\Airbnb NYC 2019 - Airbnb NYC 2019.xlsx')
```

```python
date=['2020-01-01', '2020-01-02', '2020-01-03', '2020-01-04', '2020-01-05', '2020-01-06', '2020-01-07', '2020-01-08', '2020-01-09', '2020-01-10']
price=[1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
df=pd.DataFrame({'date':date,'price':price})

plt.figure(figsize=(6,5))
plt.plot(df['date'],df['price'],linestyle='dashed',marker='x',label='Price',color='green')
plt.grid(True)
for i in range(len(df)):
    plt.text(df['date'][i],df['price'][i],df['price'][i],ha='center',va='bottom')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Price Trend')
plt.show()
```
